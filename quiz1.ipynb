```python
import pandas as pd
df = pd.read_csv('Cars.csv')
highest_hp_car = df.sort_values(by='horsepower', ascending=False).iloc[0]

print("Car with highest horsepower:")
print(f"Name: {highest_hp_car['name']}")
print(f"Horsepower: {highest_hp_car['horsepower']}")
options = ['ford mustang', 'chevrolet impala', 'pontiac grand prix', 'buick estate wagon (sw)']

print("\nHorsepower of cars in options:")
for car in options:
    car_data = df[df['name'].str.contains(car, case=False)]
    if not car_data.empty:
        max_hp = car_data['horsepower'].max()
        print(f"{car}: {max_hp} HP")
```

```python
import pandas as pd
df = pd.read_csv('Cars.csv')
count = len(df[df['mpg'] >= 35])

print(f"Number of cars with mpg >= 35: {count}")
print("\nCars with mpg >= 35:")
high_mpg_cars = df[df['mpg'] >= 35]
print(high_mpg_cars[['name', 'mpg']].sort_values(by='mpg', ascending=False))
```

```python
import pandas as pd
df = pd.read_csv('Cars.csv')
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
origin_counts = filtered_cars['origin'].value_counts()

print("Number of cars by origin (with horsepower > 100 and weight < 3000):")
print(origin_counts)
most_common_origin = origin_counts.index[0]
most_common_count = origin_counts.values[0]

print(f"\nMost common origin: {most_common_origin}")
print(f"Number of cars: {most_common_count}")
print("\nCars meeting the criteria:")
print(filtered_cars[['name', 'origin', 'horsepower', 'weight']].sort_values(by='origin'))
```

```python
import pandas as pd
df = pd.read_csv('Cars.csv')
japan_cars = df[df['origin'] == 'japan']
mean_acceleration = round(japan_cars['acceleration'].mean(), 2)

print(f"Mean acceleration of cars from Japan: {mean_acceleration}")
print("\nDetails of Japanese cars:")
print("\nNumber of Japanese cars:", len(japan_cars))
print("\nAcceleration values of Japanese cars:")
print(japan_cars[['name', 'acceleration']].sort_values(by='acceleration'))

print("\nAcceleration statistics for Japanese cars:")
print(japan_cars['acceleration'].describe())
```

```python
import pandas as pd
df = pd.read_csv('Cars.csv')
avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()
print("Average MPG by year (sorted):")
print(avg_mpg_by_year.sort_values(ascending=False))
year_highest_mpg = avg_mpg_by_year.idxmax()
highest_mpg = round(avg_mpg_by_year.max(), 2)

print(f"\nYear with highest average MPG: {year_highest_mpg}")
print(f"Average MPG in that year: {highest_mpg}")
print(f"\nDetails for cars from year {year_highest_mpg}:")
year_cars = df[df['model_year'] == year_highest_mpg]
print("\nSummary statistics for this year:")
print(year_cars['mpg'].describe())

print("\nAll cars from the year with highest average MPG:")
print(year_cars[['name', 'mpg']].sort_values(by='mpg', ascending=False))
```
